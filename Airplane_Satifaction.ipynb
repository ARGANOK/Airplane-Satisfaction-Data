```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
from matplotlib import cm
target_colors=['#8B324D','#387149']
```

```python
df = pd.read_csv('/content/train.csv')
df.head()
```

```python
df.info()
```

#Target Distribution

```python
satisfied_count = df['satisfaction'].value_counts();
```

```python
satisfied_count.plot(kind='bar', color = target_colors )
plt.xlabel('Satisfaction Type')
plt.ylabel('Satisfaction count')
plt.title('Bar graph with satisfation' ,fontsize=16, weight='bold')
# sns.countplot(x = 'satisfaction',data=df,hue='satisfaction' , palette=target_colors)
plt.show()
```

```python
plt.pie(satisfied_count, labels=satisfied_count.index, autopct='%0.1f%%', explode=(0, 0.2) , startangle=90 , colors=target_colors)
plt.title('Distribution of Passenger Satisfaction Levels', fontsize=16, weight='bold' )
```

#Genrating the data of the gender , type of travel , class , age of people

```python
gender_data = df['Gender'].value_counts().values
label1 = df['Gender'].value_counts().index
```

```python
plt.pie(gender_data, labels = label1, autopct='%0.1f%%', explode=(0, 0.2) , startangle=90 , colors=target_colors)
plt.title('Distribution of Genders of Passengers', fontsize=16, weight='bold')
plt.axis('equal')
```

```python
travel_data = df['Customer Type'].value_counts().values
label2 = df['Customer Type'].value_counts().index
```

```python
plt.pie(travel_data, labels=label2, autopct='%0.1f%%', explode=(0, 0.2) , startangle=90 , colors=target_colors)
plt.title('Distribution of Customers types', fontsize=16, weight='bold')
plt.axis('equal')
```

```python
class_data = df['Class'].value_counts().values
label3 = df['Class'].value_counts().index
```

```python
plt.pie(class_data, labels=label3, autopct='%0.1f%%', explode=(0.1, 0.1,0.1) , startangle=90 , colors=target_colors+['#324D8B'])
plt.title('Distribution of Usage of different classes', fontsize=16, weight='bold')
plt.axis('equal')
```

```python
age_kde = df['Age']
```

```python
sns.kdeplot(age_kde , fill = True)
plt.axvline(age_kde.mean(),linestyle='--',color='#324D8B',label='Average Age')
plt.title('Distribution of ages of Passengers', fontsize=16, weight='bold')
plt.legend(fontsize=10)
```

```python
ordinal_dict={5:'Excellant',
              4:'Very Good',
              3:'Good',
              2:'Bad',
              1:'Very Bad',
              0:'Not reported'}

df['Ease of Online booking'] = df['Ease of Online booking'].map(ordinal_dict)
```

```python
value1 = df['Ease of Online booking'].value_counts().values
index1 = df['Ease of Online booking'].value_counts().index
```

```python
# percentages = round(col.value_counts() / col.value_counts().sum() * 100, 2)
col = 'Ease of Online booking'
new_dif = pd.DataFrame(df[col].value_counts() , columns=[col])
new_dif['per'] = round((new_dif[col]/new_dif[col].sum())*100,2)
percent = new_dif
```

```python
sns.barplot(x = percent.index , y = 'per' , data = percent , palette = target_colors++['#324D8B'])
plt.show()
```
