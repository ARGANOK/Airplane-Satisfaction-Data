# **Project Summary -**

The dataset consists of information related to airplane satisfaction, containing various columns describing the characteristics and experiences of passengers. The dataset has a total of 103,904 entries. Let's delve into the key aspects of the dataset:

Demographic Information:

The dataset includes demographic information such as gender, age, and customer type.
The 'Gender' column indicates the gender of the passengers, while the 'Age' column provides information about their age.
The 'Customer Type' column distinguishes between different types of customers.
Travel Details:

'Type of Travel' categorizes passengers based on the purpose of their travel, distinguishing between business and personal travel.
'Class' indicates the class of service availed by passengers, such as Business, Eco, or Eco Plus.
'Flight Distance' represents the distance covered by each flight.
In-Flight Services and Satisfaction:

Various columns rate the satisfaction levels of passengers with different in-flight services, including 'Inflight wifi service,' 'Departure/Arrival time convenience,' 'Ease of Online booking,' 'Food and drink,' 'Online boarding,' 'Seat comfort,' 'Inflight entertainment,' 'On-board service,' 'Leg room service,' 'Baggage handling,' 'Checkin service,' 'Inflight service,' and 'Cleanliness.'
The 'Departure Delay in Minutes' and 'Arrival Delay in Minutes' columns provide information about the delays during departure and arrival.
Overall Satisfaction:

The 'Satisfaction' column categorizes passengers into 'Satisfied' and 'Neutral or Dissatisfied.'
Exploratory Data Analysis:

Visualizations, including count plots, box plots, bar plots, and scatter plots, have been used to explore relationships and distributions within the dataset.
Key visualizations include those related to specific services like 'Inflight service,' 'Baggage handling,' and others, along with their impact on overall satisfaction.
Insights:

The dataset allows for an exploration of factors influencing passenger satisfaction during air travel.
Analysis of in-flight services, delays, and other factors can provide insights into areas for improvement and customer preferences.
The visualizations help in identifying patterns and trends within the data, contributing to a better understanding of the factors that contribute to overall passenger satisfaction.
In summary, the dataset provides a comprehensive view of passenger experiences and satisfaction levels, offering valuable insights for airlines to enhance their services and improve overall customer satisfaction. The combination of demographic information, travel details, and in-flight service ratings enables a holistic analysis of the factors influencing the passenger experience.
