## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Statement 1 Passengers who rate "Inflight wifi service" higher are more likely to be satisfied with their overall travel experience.

Statement 2 The class of service ("Class") significantly affects the satisfaction level, with passengers in the Business class more likely to be satisfied compared to those in Economy class.

Statement 3 There is a correlation between the "Departure Delay in Minutes" and "Arrival Delay in Minutes," suggesting that flights experiencing departure delays are also likely to experience arrival delays.



### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): There is no significant difference in overall satisfaction between passengers who rate "Inflight wifi service" higher and those who rate it lower.

Alternative Hypothesis (H1): Passengers who rate "Inflight wifi service" higher are more likely to be satisfied with their overall travel experience.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import ttest_ind

wifi_satisfied = df[df['satisfaction'] == 'Satisfied']['Inflight wifi service']
wifi_not_satisfied = df[df['satisfaction'] == 'Neutral or Dissatisfied']['Inflight wifi service']

t_stat, p_value = ttest_ind(wifi_satisfied, wifi_not_satisfied, equal_var=False)

# Check the p-value
print(f"P-value for Statement 1: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in satisfaction based on Inflight wifi service.")
else:
    print("Fail to reject the null hypothesis. No significant difference in satisfaction based on Inflight wifi service.")
````

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): There is no significant difference in satisfaction levels between different classes of service.

Alternative Hypothesis (H1): Passengers in the Business class are more likely to be satisfied compared to passengers in Economy class.

````python
from scipy.stats import f_oneway

class_groups = [df[df['Class'] == cls]['satisfaction'] for cls in df['Class'].unique()]

f_stat, p_value = f_oneway(*class_groups)

# Check the p-value
print(f"P-value for Statement 2: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in satisfaction based on Class.")
else:
    print("Fail to reject the null hypothesis. No significant difference in satisfaction based on Class.")
````

To further extend our analysis, we implemented a machine learning model to predict passenger satisfaction. The RandomForestClassifier was employed for this task, and the model's performance was evaluated using accuracy, confusion matrix, and a classification report.

Throughout this process, adherence to best practices in data cleaning, preprocessing, feature selection, and model evaluation was emphasized. Additionally, considerations for ethical aspects, interpretability, and effective communication of results were highlighted.
